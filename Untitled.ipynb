```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim

# Create a dummy dataset
def create_dummy_data(num_samples, img_size):
    X = np.random.rand(num_samples, img_size, img_size, 3)  # Low-res images
    Y = np.random.rand(num_samples, img_size * 2, img_size * 2, 3)  # High-res images
    return X, Y

# Parameters
num_samples = 1000
img_size = 32
epochs = 10

# Create dummy training and testing datasets
X_train, Y_train = create_dummy_data(num_samples, img_size)
X_test, Y_test = create_dummy_data(100, img_size)

# Build a simple CNN model for super-resolution
def build_model():
    model = keras.Sequential([
        layers.Input(shape=(img_size, img_size, 3)),
        layers.Conv2D(64, (5, 5), padding='same', activation='relu'),
        layers.Conv2D(64, (5, 5), padding='same', activation='relu'),
        layers.UpSampling2D(size=(2, 2)),           # Upsampling to double the resolution
        layers.Conv2D(3, (5, 5), padding='same')    # Output layer with same size as upsampled input
    ])
    return model

model = build_model()
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, Y_train, epochs=epochs)

# Evaluate the model on the test set and calculate PSNR and SSIM
predictions = model.predict(X_test)

# Calculate PSNR and SSIM for each image in the test set
psnr_values = []
ssim_values = []

for i in range(len(X_test)):
    psnr_value = psnr(Y_test[i], predictions[i], data_range=1)  # Assuming pixel values are in [0,1]
    
    # Use a smaller window size for SSIM calculation and specify data_range
    ssim_value = ssim(Y_test[i], predictions[i], multichannel=True, win_size=3, data_range=1)
    
    psnr_values.append(psnr_value)
    ssim_values.append(ssim_value)

mean_psnr = np.mean(psnr_values)
mean_ssim = np.mean(ssim_values)

print(f'Mean PSNR: {mean_psnr:.2f} dB')
print(f'Mean SSIM: {mean_ssim:.4f}')
```
